LEGB Rule:

L: Local — Names assigned in any way within a function (def or lambda), and not declared global in that function.

E: Enclosing function locals — Names in the local scope of any and all enclosing functions (def or lambda), from inner to outer.

G: Global (module) — Names assigned at the top-level of a module file, or declared global in a def within the file.

B: Built-in (Python) — Names preassigned in the built-in names module : open, range, SyntaxError,...

##### Copying a list

```python
#### Shallow copy

import copy

# Original list with nested list
original_list = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow_copied_list = copy.copy(original_list)



print(original_list)       # Output: [[1, 2, 3], [4, 5, 6]]
print(shallow_copied_list) # Output: [[1, 2, 3], [4, 5, 6]]

print("After modifying")

# Modify the nested list in the original
original_list[0][0] = 0

print(original_list)       # Output: [[0, 2, 3], [4, 5, 6]]
print(shallow_copied_list) # Output: [[0, 2, 3], [4, 5, 6]]
```

```python
### Deep Copy
import copy

# Original list with nested list
original_list = [[1, 2, 3], [4, 5, 6]]

# Deep copy
deep_copied_list = copy.deepcopy(original_list)

# Modify the nested list in the original
original_list[0][0] = 0

print(original_list)      # Output: [[0, 2, 3], [4, 5, 6]]
print(deep_copied_list)   # Output: [[1, 2, 3], [4, 5, 6]]
```

<pre>
When to Use:
Use shallow copy if you're sure that the nested objects won't be modified or if you want changes to the nested objects 
to reflect in both the original and copied objects.

Use deep copy if you want to completely separate the original and copied objects, especially if the nested objects might be modified independently.

Understanding the distinction between shallow copy and deep copy is essential for managing mutable objects, especially in complex data structures.
</pre>

```python
class Circle:
    pi = 3.14

    # Circle gets instantiated with a radius (default is 1)
    def __init__(self, radius=1):
        self.radius = radius 
        self.area = radius * radius * Circle.pi

    # Method for resetting Radius
    def setRadius(self, new_radius):
        self.radius = new_radius
        self.area = new_radius * new_radius * self.pi

    # Method for getting Circumference
    def getCircumference(self):
        return self.radius * self.pi * 2


c = Circle()

print('Radius is: ',c.radius)
print('Area is: ',c.area)
print('Circumference is: ',c.getCircumference())
```

```python
c.setRadius(2)

print('Radius is: ',c.radius)
print('Area is: ',c.area)
print('Circumference is: ',c.getCircumference())
```

#### Polymorphism

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name+' says Woof!'
    
class Cat:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name+' says Meow!' 
    
niko = Dog('Niko')
felix = Cat('Felix')

print(niko.speak())
print(felix.speak())
```

```python
class Animal:
    def __init__(self, name):    # Constructor of the class
        self.name = name

    def speak(self):              # Abstract method, defined by convention only
        raise NotImplementedError("Subclass must implement abstract method")


class Dog(Animal):
    
    def speak(self):
        return self.name+' says Woof!'
    
    
    
class Cat(Animal):

    def speak(self):
        return self.name+' says Meow!'
    
fido = Dog('Fido')
isis = Cat('Isis')

print(fido.speak())
print(isis.speak())
```

## Error Handling

```python
def askint():
    while True:
        try:
            val = int(input("Please enter an integer: "))
        except:
            print("Looks like you did not enter an integer!")
            continue
        else:
            print("Yep that's an integer!")
            print(val)
            break
        finally:
            print("Finally, I executed!")
```

```python
## Using single except block
try:
    # Code that may raise exceptions
    x = int(input("Enter a number: "))
    result = 10 / x
    print("Result:", result)
except (ValueError, ZeroDivisionError) as e:
    # Handle multiple exceptions
    print("Error:", e)
```

```python
## Using Multiple except block

try:
    # Code that may raise exceptions
    x = int(input("Enter a number: "))
    result = 10 / x
    print("Result:", result)
except ValueError as ve:
    # Handle ValueError
    print("ValueError:", ve)
except ZeroDivisionError as ze:
    # Handle ZeroDivisionError
    print("ZeroDivisionError:", ze)
```

```python
### Handling all exception

try:
    # Code that may raise exceptions
    x = int(input("Enter a number: "))
    result = 10 / x
    print("Result:", result)
except Exception as e:
    # Handle all exceptions
    print("Error:", e)
```

## Generators

```python
# Generator function for the cube of numbers (power of 3)
def gencubes(n):
    for num in range(n):
        yield num**3
```

```python
for x in gencubes(10):
    print(x)
```

Great! Now since we have a generator function we don't have to keep track of every single cube we created.

Generators are best for calculating large sets of results (particularly in calculations that involve loops themselves) in cases where we don’t want to allocate the memory for all of the results at the same time.

## Email

```python
import smtplib

# Next is to create an STMP object that can make the method calls to log you in to your email in order to send messages.
smtp_object = smtplib.SMTP('smtp.gmail.com',587)  

# Next we run the ehlo() command which "greets" the server and establishes the connection.  250 mean success 
smtp_object.ehlo()

# When using the 587 port, this means you are using TLS encryption, which you need to initiate by running the starttls() command. 
# If you are using port 465, this means you are using SSL and you can skip this step.
smtp_object.starttls()


# For hidden passwords
import getpass

email = getpass.getpass("Enter your email: ")
password = getpass.getpass("Enter your password: ")
smtp_object.login(email,password)


# Now we can send an email using the .sendmail() method
from_address = getpass.getpass("Enter your email: ")
to_address = getpass.getpass("Enter the email of the recipient: ")
subject = input("Enter the subject line: ")
message = input("Type out the message you want to send: ")
msg = "Subject: " + subject + '\n' + message
smtp_object.sendmail(from_address,to_address,msg)

# If you get back an empty dictionary, then the sending was successful.

# now close the session
smtp_object.quit()
```

```python
import requests
import sqlite3

# Step 1: Fetch Data from FakeAPI with Authorization Token
def fetch_posts(url, token, limit=40):
    headers = {
        'Authorization': f'Bearer {token}'
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Will raise HTTPError for bad responses
        posts = response.json()
        return posts[:limit]
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return []

# Step 2: Store Data in SQLite Database with Error Handling
def store_posts_in_db(posts):
    try:
        # Connect to SQLite database (or create it if it doesn't exist)
        conn = sqlite3.connect('fakeapi_posts.db')
        cursor = conn.cursor()
        
        # Create table if it doesn't exist
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS posts (
                id INTEGER PRIMARY KEY,
                title TEXT,
                body TEXT,
                userId INTEGER
            )
        ''')
        
        # Insert posts into table
        for post in posts:
            cursor.execute('''
                INSERT INTO posts (id, title, body, userId) VALUES (?, ?, ?, ?)
            ''', (post['id'], post['title'], post['body'], post['userId']))
        
        # Commit and close connection
        conn.commit()
    except sqlite3.Error as e:
        print(f"Error storing data in database: {e}")
    finally:
        conn.close()

# Main function to execute the script
def main():
    url = "https://fakeapi.com/posts"
    token = "your_authorization_token_here"  # Replace with your actual token
    posts = fetch_posts(url, token)
    if posts:
        store_posts_in_db(posts)
        print("Data successfully fetched and stored in the database.")
    else:
        print("No data to store.")

if __name__ == "__main__":
    main()
```

```python
import requests
import sqlite3

# Configuration
BASE_URL = "https://fakeapi.com/posts"
TOKEN = "your_authorization_token_here"  # Replace with your actual token
HEADERS = {'Authorization': f'Bearer {TOKEN}'}

# Database setup
def setup_db():
    conn = sqlite3.connect('fakeapi_posts.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS posts (
            id INTEGER PRIMARY KEY,
            title TEXT,
            body TEXT,
            userId INTEGER
        )
    ''')
    conn.commit()
    conn.close()

# Create a new post
def create_post(post_data):
    try:
        response = requests.post(BASE_URL, headers=HEADERS, json=post_data)
        response.raise_for_status()
        post = response.json()
        store_post_in_db(post)
        print("Post created and stored in the database:", post)
    except requests.exceptions.RequestException as e:
        print(f"Error creating post: {e}")

# Read posts
def read_posts(limit=40):
    try:
        response = requests.get(BASE_URL, headers=HEADERS)
        response.raise_for_status()
        posts = response.json()
        store_posts_in_db(posts[:limit])
        print(f"Fetched and stored {len(posts[:limit])} posts.")
    except requests.exceptions.RequestException as e:
        print(f"Error reading posts: {e}")

# Update an existing post
def update_post(post_id, updated_data):
    try:
        response = requests.put(f"{BASE_URL}/{post_id}", headers=HEADERS, json=updated_data)
        response.raise_for_status()
        updated_post = response.json()
        update_post_in_db(post_id, updated_post)
        print("Post updated and database entry updated:", updated_post)
    except requests.exceptions.RequestException as e:
        print(f"Error updating post: {e}")

# Delete a post
def delete_post(post_id):
    try:
        response = requests.delete(f"{BASE_URL}/{post_id}", headers=HEADERS)
        response.raise_for_status()
        delete_post_in_db(post_id)
        print(f"Post with ID {post_id} deleted from API and database.")
    except requests.exceptions.RequestException as e:
        print(f"Error deleting post: {e}")

# Store a single post in the database
def store_post_in_db(post):
    try:
        conn = sqlite3.connect('fakeapi_posts.db')
        cursor = conn.cursor()
        cursor.execute('''
            INSERT INTO posts (id, title, body, userId) VALUES (?, ?, ?, ?)
        ''', (post['id'], post['title'], post['body'], post['userId']))
        conn.commit()
    except sqlite3.Error as e:
        print(f"Error storing post in database: {e}")
    finally:
        conn.close()

# Store multiple posts in the database
def store_posts_in_db(posts):
    try:
        conn = sqlite3.connect('fakeapi_posts.db')
        cursor = conn.cursor()
        for post in posts:
            cursor.execute('''
                INSERT INTO posts (id, title, body, userId) VALUES (?, ?, ?, ?)
            ''', (post['id'], post['title'], post['body'], post['userId']))
        conn.commit()
    except sqlite3.Error as e:
        print(f"Error storing posts in database: {e}")
    finally:
        conn.close()

# Update a post in the database
def update_post_in_db(post_id, updated_post):
    try:
        conn = sqlite3.connect('fakeapi_posts.db')
        cursor = conn.cursor()
        cursor.execute('''
            UPDATE posts SET title = ?, body = ?, userId = ? WHERE id = ?
        ''', (updated_post['title'], updated_post['body'], updated_post['userId'], post_id))
        conn.commit()
    except sqlite3.Error as e:
        print(f"Error updating post in database: {e}")
    finally:
        conn.close()

# Delete a post from the database
def delete_post_in_db(post_id):
    try:
        conn = sqlite3.connect('fakeapi_posts.db')
        cursor = conn.cursor()
        cursor.execute('DELETE FROM posts WHERE id = ?', (post_id,))
        conn.commit()
    except sqlite3.Error as e:
        print(f"Error deleting post from database: {e}")
    finally:
        conn.close()

# Main function to execute the script
def main():
    setup_db()
    
    # Example data for CRUD operations
    new_post = {"title": "New Post", "body": "This is a new post", "userId": 1}
    updated_post_data = {"title": "Updated Post", "body": "This is an updated post", "userId": 1}
    post_id_to_update = 1
    post_id_to_delete = 2
    
    # Create a new post
    create_post(new_post)
    
    # Read posts
    read_posts()
    
    # Update an existing post
    update_post(post_id_to_update, updated_post_data)
    
    # Delete a post
    delete_post(post_id_to_delete)

if __name__ == "__main__":
    main()
```
